Ноутбук для управления парсером

### Импорты

```python
import os
import json
from dotenv import load_dotenv

from src.parser.control_parser import start_parser


# Загрузка переменных окружения из файла .env
load_dotenv()
```

### Настройка парсера

```python
try:
    EMAIL = os.getenv("EMAIL") # NCBI требует указания email для доступа к API
except:
    EMAIL = input("Enter your email for NCBI API access: ")

try:
    NCBI_API_KEY = os.getenv("NCBI_API_KEY") # NCBI API Key для увеличения лимитов запросов
except:
    NCBI_API_KEY = input("Enter your NCBI API Key (or leave blank to skip): ")


query = "(Alzheimer's disease[Title/Abstract]) AND " +\
    "(therapeutic targets[Title/Abstract] OR " +\
    "drug targets[Title/Abstract])"
    
limit = 100 # Требуемое количество статей
```

### Запуск парсера

```python
start_parser(
    base_query=query,
    limit=limit,
    email=EMAIL,
    api_key=NCBI_API_KEY
)
```
